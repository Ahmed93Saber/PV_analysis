```python
import cv2

asc_20_path = r"C:\Users\ahmed\OneDrive\Desktop\tiff_imgs\EL_Increment\4_251103_EL_M8_40pro_6000ms_20mA_PM045104_242.tif"
asc_20_img = cv2.imread(asc_20_path, -1)

des_20_path = r"C:\Users\ahmed\OneDrive\Desktop\tiff_imgs\EL_Increment\14_251103_EL_M8_40pro_6000ms_20mA_PM053304_890.tif"
des_20_img = cv2.imread(des_20_path, -1)

# calculate and ratio and plot
ratio_20 = cv2.divide(des_20_img,asc_20_img, scale=1.0)
cv2.imwrite(r"C:\Users\ahmed\OneDrive\Desktop\tiff_imgs\EL_Increment\ratio_20.tif", ratio_20)

#normalize the ratio image for better visualization
ratio_20 = cv2.normalize(ratio_20, None, 0, 255, cv2.NORM_MINMAX)
ratio_20 = ratio_20.astype('uint8')
# plot as a colormap
ratio_20 = cv2.applyColorMap(ratio_20, cv2.COLORMAP_JET)


# display the images
cv2.imshow("Asc 20mA", asc_20_img)
cv2.imshow("Des 20mA", des_20_img)
cv2.imshow("Ratio 20mA", ratio_20)
cv2.waitKey(0)

```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import cv2
from matplotlib.colors import LogNorm
from PIL import Image

def map_EL_div(image1, image2, title):
    #x plays a role in distributing the values on the color range
    # vi is vmax
    #v=70000
    if image1.shape != image2.shape:
        image2=cv2.resize(image2,(image1.shape[1],image1.shape[0]))




    image1_safe = np.where(image1 == 0, np.nan, image1)
    normfordiv=mcolors.LogNorm(vmin=0.1, vmax=10)
    image=image2/image1_safe
    im=plt.imshow(image,cmap="seismic",norm=normfordiv)
    plt.title(title, fontsize=20, pad=10)
    plt.xticks([])
    plt.yticks([])
    cbar=plt.colorbar(im)
    #cbar.set_ticks([100, 500, 1000,2000, 4000,10000, 30000, 50000, 60000, 0.39])
    cbar.ax.set_title("ratio", fontsize=14, pad=20)
    #cbar.ax.yaxis.set_label_position("left")
    #plt.tight_layout()
    plt.show()

asc_20_path = r"C:\Users\ahmed\OneDrive\Desktop\tiff_imgs\EL_Increment\4_251103_EL_M8_40pro_6000ms_20mA_PM045104_242.tif"
asc_20_img = Image.open(asc_20_path)

des_20_path = r"C:\Users\ahmed\OneDrive\Desktop\tiff_imgs\EL_Increment\14_251103_EL_M8_40pro_6000ms_20mA_PM053304_890.tif"
des_20_img = Image.open(des_20_path)

map_EL_div(np.array(des_20_img), np.array(asc_20_img), "20 mA Ratio Image")
```
